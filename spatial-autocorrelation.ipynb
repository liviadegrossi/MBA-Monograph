```python
import geopandas as gpd
import matplotlib.pyplot as plt
import libpysal as lps
# import esda
```

```python
# reading the hexagonal grid files
grid_30km = gpd.read_file('data/30km/grid-30km.shp')
grid_50km = gpd.read_file('data/50km/grid-50km.shp')
grid_90km = gpd.read_file('data/90km/grid-90km.shp')
```

```python
grid_30km.head()
```

## Spatial similarity: Queen contiguity

```python
# defining the neighbors of the 30 km hexagonal grids
neighbors_30km = lps.weights.Queen.from_dataframe(grid_30km)
```

```python
# plotting the neighbors of the 30 km hexagonal grids
ax = grid_30km.plot(edgecolor="grey", facecolor="w")
f, ax = neighbors_30km.plot(
    grid_30km, 
    ax=ax, 
    edge_kws=dict(color="r", linestyle=":", linewidth=1),
    node_kws=dict(marker=""),
)
ax.set_axis_off()
plt.savefig('images/neighbors-30km.png')
```

```python
# defining the neighbors of the 50 km hexagonal grids
neighbors_50km = lps.weights.Queen.from_dataframe(grid_50km)
```

```python
# plotting the neighbors of the 50 km hexagonal grids
ax_50km = grid_50km.plot(edgecolor="grey", facecolor="w")
f, ax = neighbors_50km.plot(
    grid_50km, 
    ax=ax_50km, 
    edge_kws=dict(color="r", linestyle=":", linewidth=1),
    node_kws=dict(marker=""),
)
# ax.set_axis_off()
ax_50km.set_axis_off()
plt.savefig('images/neighbors-50km.png')
```

```python
# defining the neighbors of the 90 km hexagonal grids
neighbors_90km = lps.weights.Queen.from_dataframe(grid_90km)
```

```python
# plotting the neighbors of the 90 km hexagonal grids
ax_90km = grid_90km.plot(edgecolor="grey", facecolor="w")
f, ax = neighbors_90km.plot(
    grid_90km, 
    ax=ax_90km, 
    edge_kws=dict(color="r", linestyle=":", linewidth=1),
    node_kws=dict(marker=""),
)
# ax.set_axis_off()
ax_90km.set_title("Neighborhood: Queen contiguity")
ax_90km.set_axis_off()
plt.savefig('images/neighbors-90km.png')
```
